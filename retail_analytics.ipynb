```python
import pandas as pd

# Load the dataset
df = pd.read_csv("retail_data.csv")

# Add total amount column
df['TotalAmount'] = df['Quantity'] * df['Price']

# Summary
print("üîπ Basic Info:")
print(df.info())
print("\nüîπ First 5 Rows:\n", df.head())

# Total sales
total_sales = df['TotalAmount'].sum()
print(f"\nüí∞ Total Sales: ‚Çπ{total_sales}")

# Sales by category
category_sales = df.groupby("Category")["TotalAmount"].sum().sort_values(ascending=False)
print("\nüìä Sales by Category:\n", category_sales)

# Top 5 products by revenue
top_products = df.groupby("Product")["TotalAmount"].sum().sort_values(ascending=False).head(5)
print("\nüèÜ Top 5 Products by Revenue:\n", top_products)

# Unique customers
unique_customers = df['CustomerID'].nunique()
print(f"\nüë• Total Unique Customers: {unique_customers}")
```
